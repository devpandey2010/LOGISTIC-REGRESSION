```python
#iris dataset based petal length ,petal width,sepal length,sepal width we have to predict the type of iris flower
#iris flower are of three types 1.Setosa 2.versicolour 3.virginica
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris.DESCR
```

```python
iris.target
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df["target"]=iris.target
```

```python
df
```

```python
df.info()
```

```python
#0>>setosa 1>>versicolour 2>>viginica
```

```python
#we will see now only for binary classification so we will need only two class
#we will ignore class 2
df=df[df["target"]!=2]
df
```

```python
df.target.unique()
```

```python
#split into x and y
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
#model training
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)

```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.predict_proba(x_test) #corresponding to higher probability class is predicted
```

```python
#model evaluation
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,classification_report
print(classification_report(y_test,y_pred))
print(f"accuracy score:{accuracy_score(y_test,y_pred)}")
print(f"confusion matrix:{confusion_matrix(y_test,y_pred)}")
print(f"precision score:{precision_score(y_test,y_pred,average='micro')}")
print(f"recall score:{recall_score(y_test,y_pred,average='micro')}")
```

```python
#first two rows ,0and 1>>precision,recall,f1 score and no of data points with respect to respective class
#avg  preacision of class  0 and class 1,similarly for recall and f_score
#weighted avg is used in class imbalance,it gives importance to performance of dominant class
```

```python
from sklearn.metrics import roc_curve,auc

#probability of class 1 only
model.predict_proba(x_test)[:,1]
```

```python
y_pred_proba = model.predict_proba(x_test)[:,1]
y_pred_proba
```

```python
#reciever operating characteristic curve
#input:y_test and y_pred_proba it will return TPR,FPR,with different cutoff of probabilities

fpr,tpr,threshold=roc_curve(y_test,y_pred_proba)
```

```python
threshold
```

```python
fpr
```

```python
tpr
```

```python
auc(fpr,tpr)
```

```python
#plot roc_auc curve
from sklearn.metrics import roc_auc_score
roc_auc=roc_auc_score(y_test,y_pred_proba)
plt.figure(figsize=(10,7))#create a new figure with a specified size
plt.plot(fpr,tpr,color='orange',lw=2,label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0,1],[0,1],color='darkblue',lw=2,linestyle='--')#plot the diagonal line representing random guessing
plt.xlim([0.0,1.0]) #set x-axis limits
plt.ylim([0.0,1.0]) #set y-axis limits
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
